Convertir image en echelle de gris

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline

test_image = cv2.imread('test12.jpg')

haar_cascade_face = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_alt2.xml')

test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)

plt.imshow(test_image_gray, cmap='gray')
```

Detection de visage

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5);

print('Faces found: ', len(faces_rects))
def convertToRGB(image): 
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.imshow(convertToRGB(test_image))
```

Definir la fonction

```python
def detect_faces(cascade, image,neighbors, scaleFactor=1.1):

    image_copy = image.copy()
   
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
    
    
    faces_rects = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=neighbors,flags = 0)
    print('Faces found: ', len(faces_rects))
    for (x, y, w, h) in faces_rects:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
    plt.imshow(convertToRGB(image_copy))
```

```python
test_image2 = cv2.imread('baby3.png')
faces = detect_faces(haar_cascade_face,test_image2,5)
```

Fonction finale avec enregistrement

```python
def detect_faces(cascade, image,neighbors, scaleFactor=1.1):

    image_copy = image.copy()
   
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
    
    
    faces_rects = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=neighbors,flags = 0)
    print('Faces found: ', len(faces_rects))
    for (x, y, w, h) in faces_rects:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
    plt.imshow(convertToRGB(image_copy))
    resultat = "resultat.jpg"
    
    cv2.imwrite(resultat,image_copy)
    #l'image sera enregristrée dans le repertoire content

faces = detect_faces(haar_cascade_face,test_image3,1)
```
